```python
headers = {'User-Agent': 'UNC Journo Class'}
base_url = 'https://www.metmuseum.org/art/collection/'
url = base_url + 'search/198560?sortBy=Relevance&amp;ao=on&amp;ft=*&amp;offset=0&amp;rpp=20&amp;pos=2'
resp = requests.get(url, headers=headers)
```

```python
body_str = resp.content.decode('utf-8')
body_str
```

```python
sel = scrapy.Selector(text=body_str)
sel
```

```python
#table = sel.css('collection-details__tombstone')[0]
table = sel.css('dl')
```

```python
table
```

```python
cols = table.css('dt').xpath('string()').extract()
```

```python
cols
```

```python
title = sel.css('h1').xpath('string()').extract()
title
```

```python
rows = table.css('dd').xpath('string()').extract()
rows
```

```python
paintings = [] 
for r in rows:
    data = {}
    for i, d in enumerate(r.css('dd')):
        a = d.css('a')
        if a:
            t = a.xpath('text()').extract()[0]
            data['href'] = a.xpath('@href').extract()[0]
        else:
            t = d.xpath('text()').extract()[0]
        data[cols[i]] = t
    paintings.append(data)
```
